Iris Data Classification

Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
# %matplotlib inline 
```

Step 2: Load Dataset

```python
df = pd.read_csv('./iris.data')
df.head()
```

```python
columns = ['sepal length','sepal width','petal length','petal width','class labels']
```

```python
df.describe()
```

```python
df = pd.read_csv('./iris.data', names=columns)
df.head()
```

Step 3: Describe the dataset

```python
sns.pairplot(df, hue='class labels')
```

Step 4: Seperate x and y

```python
data = df.values
x = data[:,0:4]
y = data[:,4]
```

Step 5: Split train and test data

```python
xtrain, xtest, ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=50)
```

Step 6: Model is taken as Support Vector Machine 

```python
svc_model = SVC()
svc_model.fit(xtrain,ytrain)
```

```python
svc_ypred = svc_model.predict(xtest)
```

Step 7: Test Accuracy

```python
score_svc = accuracy_score(ytest,svc_ypred)
score_svc
```

Step 8: Using Logistic Regression Model

```python
lr_model = LogisticRegression()
lr_model.fit(xtrain,ytrain)
```

Print accuracy

```python
lr_ypred = lr_model.predict(xtest)
score_lr = accuracy_score(ytest,lr_ypred)
print(score_lr)
```

Step 9: Using Decision Tree Classifier

```python
model_dtc = DecisionTreeClassifier()
model_dtc.fit(xtrain,ytrain)
```

```python
dtc_ypred = model_dtc.predict(xtest)
score_dtc = accuracy_score(ytest,dtc_ypred)
score_dtc
```
